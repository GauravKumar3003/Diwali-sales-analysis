```python
# import python libraries

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
# Shape of the csv file(totl rows and columns)(1)
df.shape
```

```python
rows,cols=df.shape
print("Number of rows:",{rows})
print("Number of columns:",{cols})
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values(5)
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
# show all columns(6)
df.columns
```

```python
#rename column (from marital_status to shaadi)(7)
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
# Total profit if all amount(2)
total_profit = df["Amount"].sum()
print("Total Profit:",total_profit)
```

```python
# Maximum purchase(3)
max_purchase1=df["Amount"].max()
max_purchase2=df["Orders"].max()
print("Max purchase by amount",max_purchase1)
print("Max purchase by orders",max_purchase2)
```

```python
# Minimum purchase(4)
min_purchase1=df["Amount"].min()
min_purchase2=df["Orders"].min()
print("Min purchase by amount",min_purchase1)
print("Min purchase by orders",min_purchase2)
```

```python
# Average of all amount(8)
avg = df["Amount"].mean()
print("Average amount:",avg)
```

```python
# Total user_id that have the orders3
user=df[df["Orders"]==3]
user1=user["User_ID"].tolist()
print(user1)
```

```python
# Total numbers of users have orders3
length_orders3=len(user)
print(length_orders3)
```

```python
# Count of all orders(9)
num_orders = len(df["Orders"].unique())
print("Number of orders:",num_orders)
```

```python
# Mean of the orders
mean=df["Orders"].mean()
print(mean)
```

```python
mode=df["Orders"].mode()
print(mode)
```

```python
median=df["Orders"].median()
print(median)
```

```python
# Standard deviation
std=df["Orders"].std()
print(std)
```

```python
variance=df["Orders"].var()
print(variance)
```

##### Exploratory Data analysis (Chart representation)

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for age group and it's count
ax = sns.countplot(x = 'Age Group',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plotting a pie chart for gender and it label
count=df["Gender"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(count.values,labels=count.index,autopct='%1.1f%%')# autopct(use for percentage)
plt.show()  
```

```python
# Plotting a pie chart for age group and it label
count1=df["Age Group"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(count1.values,labels=count1.index,autopct='%1.1f%%')# autopct(use for percentage)
plt.show()  
```

```python
# Plotting a line graph foe gender
counts=df.groupby("Gender").size()
counts.plot(kind="line")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Plotting a line graph foe age group
counts1=df.groupby("Age Group").size()
counts1.plot(kind="line")
plt.xlabel("Age group")
plt.ylabel("Count")
plt.show()
```

```python
# Compare of gender with age group using bar graph(10)
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')# hue(use for combining)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# State wise orders(11)
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# occupation wise orders
sales_state1 = df.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state1, x = 'Occupation',y= 'Orders')
```

```python
# Product wise orders
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x = 'Age Group',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state1 = df.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data = sales_state1, x = 'Occupation',y= 'Orders')
for bars in ax.containers:
    ax.bar_label(bars)
    
```
